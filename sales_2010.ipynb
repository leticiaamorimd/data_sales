```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
purchase_eu_df = pd.read_csv('purchase_europe.csv')
sales_europe_df = pd.read_csv('sales_europe.csv')
sales_middle_df = pd.read_csv('sales_middle_East.csv')
sales_north_df = pd.read_csv('sales_north_america.csv')

#sales_europe_df 
#sales_middle_df
```

```python
#II - EUROPE
sales_europe_df.plot.line(x='country' , y='sales_figure', marker='d')
plt.title('Europe - 2010')
plt.xlabel('Countries')
plt.ylabel('Sales figure')
plt.legend(loc='lower left')

plt.savefig('europe_sales.jpeg')
```

```python
#II - MIDDLE EAST
sales_middle_df.plot.line(x='country' , y='sales_figure', marker='d')
plt.title('Middle East Sales - 2010')
plt.xlabel('Countries')
plt.ylabel('Sales figure')
plt.legend(loc='lower left')

plt.savefig('sales_middle.jpeg ')
```

```python
#II - 
sales_north_df.plot.line(x='country' , y='sales_figure', marker='d')
plt.title('North America - 2010')
plt.xlabel('Countries')
plt.ylabel('Sales figure')
plt.legend(loc='lower left')

plt.savefig('sales_north.jpeg')
```

```python
#i. Sales for the year of 2010 across all countries

#sales_north_df
#sales_europe_df
#sales_middle_df

merge_two = pd.merge(sales_middle_df, sales_europe_df, how='outer')
merge_all = pd.merge(merge_two, sales_north_df, how='outer')

merge_all
```

```python
# iii. Using line charts Purchase vs Sales for the year of 2010 for all countries in Europe that have
# both sales and purchase records (one chart per country)

sales_europe_df
purchase_eu_df

#merge all dataframe 
europe_df = pd.merge(sales_europe_df, purchase_eu_df)

#isin - to remain only the rows that have a match in purchase_eu 
all_df = europe_df[sales_europe_df.country.isin (purchase_eu_df.country)]

all_df
```

```python
#iv - Perform task iii above using plotly
all_df.plot()
plt.title('Germany - Sales and Purchase - 2010')
plt.xlabel('Months')
plt.ylabel('Sales figure')
plt.legend(loc='upper left')

plt.savefig('germany_sales_purchase.jpeg')


#how delete year and month lines from graph?
```
